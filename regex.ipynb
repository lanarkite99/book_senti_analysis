# Load the file

```python
with open("miracle_in_the_andes.txt","r",encoding="UTF-8") as file:
    book=file.read()
```

# calculate no of chapters

```python
import re
```

```python
pattern=re.compile("Chapter [0-9]")
num=re.findall(pattern,book)
len(num)
```

# list the sentences where word "love" is used

```python
pattern=re.compile("[^.]* love[^a-zA-Z]+[^.]*.")
sent=re.findall(pattern,book)
len(sent)
```

# count the most used word

```python
pattern=re.compile("[a-zA-Z]+")
res=re.findall(pattern,book.lower())
res[:10]
```

```python
dict={}
for word in res:
    if word in dict.keys(): 
        dict[word]=dict[word]+1
    else:
        dict[word]=1
```

```python
word_list=[(key,value) for (value,key) in dict.items()]
sorted(word_list)
```

# function that finds the occurence of any word

```python
def find(w):
    pattern=re.compile("[a-zA-Z]+")
    res=re.findall(pattern,book.lower())
    dict={}
    for word in res:
        if word in dict.keys(): 
            dict[word]=dict[word]+1
        else:
            dict[word]=1
    try:
        return dict[w]
    except:
        return "word not found"
```

# call the function

```python
find("love")
```
